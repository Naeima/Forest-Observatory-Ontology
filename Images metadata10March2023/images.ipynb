```python
for index, row in df.iterrows():
    g.add((URIRef(ID+row['ID']), RDF.type, SOSA.Observation)) 
    g.add((URIRef(ID+row['ID']), foo.ID, Literal(row['ID'], datatype=XSD.string))) 
    
    g.add((URIRef(schema+'Images.forest-observatory.cardiff.ac.uk/DGFC#'), RDF.type,  SOSA.Sensor)) 
    g.add((URIRef(ID+row['ID']), URIRef(schema+'Images.forest-observatory.cardiff.ac.uk/DGFC#'), Literal(row['ID'], datatype=XSD.string)))    
    
    g.add((URIRef(foo+'path'), RDF.type, SOSA.observedProperty))
    g.add((URIRef(ID+row['ID']), foo.path, Literal(row['path'], datatype=XSD.anyURI)))
    
    g.add((URIRef(foo+'name'), RDF.type, SOSA.observedProperty))
    g.add((URIRef(ID+row['ID']), foo.name, Literal(row['name'], datatype=XSD.string)))  
    
    g.add((URIRef(foo+'make'), RDF.type, SOSA.observedProperty))
    g.add((URIRef(ID+row['ID']), foo.make, Literal(row['make'], datatype=XSD.string))) 
    
    g.add((URIRef(foo+'model'), RDF.type, SOSA.observedProperty))
    g.add((URIRef(ID+row['ID']), foo.model, Literal(row['model'], datatype=XSD.string)))
    
    g.add((URIRef(foo+'date'), RDF.type, SOSA.observedProperty))
    g.add((URIRef(ID+row['ID']), foo.date, Literal(row['date'], datatype=XSD.date)))
    
    g.add((URIRef(foo+'time'), RDF.type, SOSA.observedProperty))
    g.add((URIRef(ID+row['ID']), foo.time, Literal(row['time'], datatype=XSD.time)))
```

```python
print(g.serialize(format='turtle'))
```

```python
# saving ontology to disk
g.serialize('Image.rdf', format="ttl")
```
