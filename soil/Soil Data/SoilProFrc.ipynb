```python
from rdflib import Graph, Literal, RDF, URIRef, Namespace 
from rdflib.namespace import FOAF , XSD, SSN, SOSA 
import urllib.parse
import pandas as pd 
import matplotlib.pylab as plt
from matplotlib import pyplot
```

```python
df = pd.read_csv('SoilProFrc.csv')
```

```python
df.tail(2)
```

```python
g = Graph()
ID = Namespace('foo_')
owl = Namespace('http://www.w3.org/2002/07/owl#')
sosa = Namespace('http://www.w3.org/ns/sosa/')
foo = Namespace('http://www.ontology/ns/foo/1.1#')
lat = Namespace('http://www.w3.org/2003/01/geo/wgs84_pos#')
long =Namespace('http://www.w3.org/2003/01/geo/wgs84_pos#')
alt = Namespace('http://www.w3.org/2003/01/geo/wgs84_pos#')
UNIT= Namespace('http://qudt.org/vocab/unit')
schema = Namespace('http://schema.org/')
XSD= Namespace('http://www.w3.org/2001/XMLSchema#')
```

```python
for index, row in df.iterrows():
    
    g.add((URIRef(ID+row['Identifier']), RDF.type, sosa.Observation))
    g.add((URIRef(ID+row['Identifier']), foo.ID, Literal(row['Identifier'], datatype=XSD.string))) 
    
    g.add((URIRef(schema+'Soil_Properties_FRC'), RDF.type,  sosa.Sensor))
    g.add((URIRef(ID+row['Identifier']), URIRef(schema+'Soil_Properties_FRC'), Literal(row['Identifier'], datatype=XSD.string))) 
   
    g.add((URIRef(foo+'Site'), RDF.type, sosa.FeatureOfInterest))
    g.add((URIRef(ID+row['Identifier']), foo.Site, Literal(row['Site'], datatype=XSD.string)))
    
    g.add((URIRef(foo+'Land_Use'), RDF.type, sosa.observedProperty))
    g.add((URIRef(ID+row['Identifier']), foo.Land_Use, Literal(row['Land_Use'], datatype=XSD.string)))

    g.add((URIRef(foo+'Plot_Name'), RDF.type, sosa.observedProperty))
    g.add((URIRef(ID+row['Identifier']), foo.Plot_Name, Literal(row['Plot_Name'], datatype=XSD.string)))

    g.add((URIRef(foo+'Subplot'), RDF.type, sosa.observedProperty))
    g.add((URIRef(ID+row['Identifier']), foo.Subplot, Literal(row['Subplot'], datatype=XSD.integer))) 

    g.add((URIRef(foo+'Horizon'), RDF.type, sosa.observedProperty))
    g.add((URIRef(ID+row['Identifier']), foo.Horizon, Literal(row['Horizon'], datatype=XSD.string))) 

    g.add((URIRef(foo+'Soil_pH'), RDF.type, sosa.observedProperty))
    g.add((URIRef(ID+row['Identifier']), foo.Soil_pH, Literal(row['Soil_pH'], datatype=XSD.float)))

    g.add((URIRef(foo+'Total_C'), RDF.type, sosa.observedProperty))
    g.add((URIRef(ID+row['Identifier']), foo.Total_C, Literal(row['Total_C'], datatype=XSD.float)))

    g.add((URIRef(foo+'Total_N'), RDF.type, sosa.observedProperty))
    g.add((URIRef(ID+row['Identifier']), foo.Total_N, Literal(row['Total_N'], datatype=XSD.float)))

    g.add((URIRef(foo+'Total_P'), RDF.type, sosa.observedProperty))
    g.add((URIRef(ID+row['Identifier']), foo.Total_P, Literal(row['Total_P'], datatype=XSD.float)))

    g.add((URIRef(foo+'inorganic_P'), RDF.type, sosa.observedProperty))
    g.add((URIRef(ID+row['Identifier']), foo.inorganic_P, Literal(row['inorganic_P'], datatype=XSD.float)))

    g.add((URIRef(foo+'Sand'), RDF.type, sosa.observedProperty))
    g.add((URIRef(ID+row['Identifier']), foo.Sand, Literal(row['Sand'], datatype=XSD.float)))

    g.add((URIRef(foo+'Silt'), RDF.type, sosa.observedProperty))
    g.add((URIRef(ID+row['Identifier']), foo.Silt, Literal(row['Silt'], datatype=XSD.float)))

    g.add((URIRef(foo+'Clay'), RDF.type, sosa.observedProperty))      
    g.add((URIRef(ID+row['Identifier']), foo.Clay, Literal(row['Clay'], datatype=XSD.float)))
```

```python
print(g.serialize(format='turtle'))
```

```python
# saving ontology to disk
g.serialize('SoilProFrc.rdf', format='ttl')
```
